```python
import tensorflow as tf
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.options.mode.copy_on_write = True
```

```python
url = 'https://drive.google.com/file/d/1C5AkLrXjUu-BlhgzZdIIqZ-paavDuYkt/view?usp=sharing'
path = 'https://drive.google.com/uc?export=download&confirm=1&id='+url.split('/')[-2]
df = pd.read_csv(path)
```

```python
#see z-score normalization, better for outlier detection
def znorm(column):
    df[column] = (df[column] - df[column].mean()) / df[column].std()

#normal absolute max normalization
def maxnorm(column):
    df[column] = df[column] / df[column].abs().max()


csv_path = tf.keras.utils.get_file(
    origin='https://drive.google.com/uc?export=download&id=1OsehlRnbGm1WMDcxRJ3o6ClwRp0X9Js-') #node 2 7/19/24
olddf = pd.read_csv(csv_path)
df = pd.DataFrame() #empty dataframe
columns = ['Temperature', 'Pressure', 'Humidity', '1.0 std', '2.5 std', '10 std']

for i in range(6):
    df[columns[i]] = olddf.loc[:, olddf.columns[i]] #standardizes naming

df['Temperature'] = (df['Temperature'] / 500) + 24 #converts to deg C

oldcolumns = list(map(str.lower,list(olddf.columns))) #some files don't have time so these are optional, we might not even use time as an input
if 'date' in oldcolumns:
    df['Date'] = olddf.loc[:, olddf.columns[oldcolumns.index('date')]]
    df['Time'] = olddf.loc[:, olddf.columns[oldcolumns.index('time')]]

#no longer uses df.iloc which caused a bunch of errors woo!

df['Fire'] = 0 #empty column

#enter the intervals here as a list, eg 194-251, 398-438 would be [194, 251, 398, 438]
intervals = [198, 249, 404, 461, 855, 931, 1090, 1119, 1280, 1338, 1541, 1589] 

for i in range(0,len(intervals), 2):
    x = range(intervals[i], intervals[i+1])
    df.loc[x, 'Fire'] = 1 #in each range, make the value in Fire column a 1 instead of a 0

plt.plot(df.index[:], df['Fire']) #graph to see intervals
```

```python
while 0 in df.loc[0, df.columns[0:6]].values: #while theres a 0 in the first row in any of the first 6 columns:
    df.drop(index=0, inplace=True) #drops that row
    df.reset_index(drop=True, inplace=True) #fixes index

df.drop_duplicates(subset=df.columns[0:6],ignore_index=True, inplace=True) #drop dups actually works now
df = df.assign(Humidity=df['Humidity'].abs()) #turns humidity positive since its just negative sometimes
df.drop(df[df['Humidity'] == 0].index, inplace=True) #drops anytime humidity is 0
df.drop(df[df['Pressure'] == 0].index, inplace=True) 

for i in range(6):
    column = str(df.columns[i])
    df[column] = df[column].astype(float)

znorm("1.0 std")
znorm("2.5 std")
znorm("10 std")

for i in range(6):
    column = str(df.columns[i])
    maxnorm(column)





path=r'C:\Users\luciu\Desktop\Processed CSVs'
df.to_csv(os.path.join(path,r'Preprocessed_node_2_07_19_24.csv'), index=False)


#plt.plot(df.index[:5000], df['Temperature'][:5000])
#plt.plot(df.index[:], df['Fire'], df['2.5 std'])
df
```
